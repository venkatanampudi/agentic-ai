```python
# Converting "Value" column data type from float64 to int

df['Value_New']=df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
# Applying a function on value column (with custom lambda function)

df['New Value'] =  df['Value'].apply(lambda x:x*2)
df.head()
```

```python
# Data aggregating and grouping by categorical column "Product"

grouped_mean=df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
# Data aggregating (sum) and grouping by two categorical columns "Product" & "Region"

grouped_sum=df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

```python
# Data aggregating (mean) and grouping by two categorical columns "Product" & "Region"


grouped_mean_two=df.groupby(['Product','Region'])['Value'].mean()
print(grouped_mean_two)
```

```python
# Aggregating multiple functions 

grouped_agg=df.groupby('Region')['Value'].agg(['mean','sum','count'])
print(grouped_agg)
```

```python
# Merging and joining dataframes 

# Creating sample dataframes

df1=pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'Key':['A','B','D'],'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
# Merge datafram on the Key columns

pd.merge(df1,df2,on="Key",how="inner")
```

```python
# outer join --> both dataframes will be joined 

pd.merge(df1,df2,on="Key",how="outer")
```

```python
# leftouter join (means more priority on df1)

pd.merge(df1,df2,on="Key",how="left")
```

```python
# rightouter join (means more priority on df2)
pd.merge(df1,df2,on="Key",how="right")
```
