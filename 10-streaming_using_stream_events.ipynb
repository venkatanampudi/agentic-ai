### Streaming 
##### Methods: stream() and astream()
* These methods are sync and async methods for streaming back results.

##### Additional parameters in streaming modes for graph state.
* values: This streams the full state of the graph after each node is called
* updates: This stream updates to the state of the graph after node is called



```python
### after creating the necessary code before this cell, create a thread using config and stream_mode="udpates"
config={"configurable":{"thread_id":"1"}}

for chunk in graph_builder.stream({"messages":"Hi, My name is Venkat and I like to play Badminton"},config,stream_mode="updates"):
    print(chunk)
```
