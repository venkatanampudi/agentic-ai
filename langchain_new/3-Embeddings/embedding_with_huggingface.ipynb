### Sentence Transformers on Hugging Face 
##### Hugging Face sentence-transformers is a Python framework for state-of-the-art sentence, text and image embeddings. One of the embedding models is used in the HuggingFaceEmbeddings class. We have also added an alias for SentenceTransformerEmbeddings for users who are more familiar with directly using that package.
* pip install sentence-transformers
* pip install langchain_huggingface

```python

from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text="This is my test document"
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
### even we can pass the multiple documents using embed_documents() method using list
text1="Machine learning models can analyze large datasets to find patterns."
text2="AI systems help discover meaningful insights from vast amounts of data."
doc_result=embeddings.embed_documents([text1,text2])
doc_result[0]
```
