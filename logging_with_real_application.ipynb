```python
import logging 

# Basic config settings
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    # loggingtime-loggername(ex:root)-severitylevel-actual log message (all in string(s) format) 
    datefmt='%Y-%m-%d %H:%M:%S',
    #YYYY-MM-DD HH:MM:SS (25-12-16 17:05:52) format
    handlers=[
        logging.FileHandler("app1.log"),
        logging.StreamHandler()
        ]
)

logger=logging.getLogger("ArithmeticApplication")

# adding two numbers 
def add(a,b):
    result=a+b
    logger.debug(f"Adding {a} + {b} == {result}")
    return result

# subtracting two numbers 
def subtract(a,b):
    result=a-b
    logger.debug(f"Subtracting {a} - {b} == {result}")
    return result

# Multiplying two numbers 
def multiply(a,b):
    result=a*b
    logger.debug(f"Multiplying {a} * {b} == {result}")
    return result

# Dividing two numbers 
def divide(a,b):
    try:
        result=a / b
        logger.debug(f"Dividing {a} / {b} == {result}")
        return result
    except ZeroDivisionError:
        logger.error("Division by Zero error raised")
        return None
    
# calling functions 
   
add(10,15)
subtract(10,15)
multiply(10,15)
divide(10,0)
```
