### UV Package Manager
``` path= /mnt/c/projects/agentic-ai/langchain-updated```
``` step 1.go to path and uv init ==> once this is executed, you can see project initialization with necessary folders```
``` step 2.create virtual environment using  command "uv venv" & this will create virtual env called ".venv" (u must see ."venv" folder inside your path)``
``` step 3.activate the venv using "source .venv/bin/activate" ```
``` step 4. create "requriements.txt" file outside ".venv" folder to install the necessary libraries```
``` step 5. install the requirements.txt file using command "uv add -r requirements.txt" ```
``` step 6. acquire the keys: google api key,groq api key,open api key,huggingface api key, etc., etc keys```
``` if any independent libraries need to install use command "uv add <<library-name>>" ```
