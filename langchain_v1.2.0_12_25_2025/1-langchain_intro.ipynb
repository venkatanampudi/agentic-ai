### Creating a basic (autonomous) agent (using a tool)

```python
from langchain.agents import create_agent

# Define a function for an example
def get_weather(city:str)->str:
    """Get the weather for a city"""
    return f"The weather in {city} is sunny."

agent=create_agent(
    model="gpt-5",
    tools=[get_weather],
    system_prompt="You are an helpful assistant"
)

# after passing this info, tool will get added as "get_weather" 
```

```python
agent
```

```python
### Run the agent 
response=agent.invoke({"messages":[{"role":"user","content":"What is weather in New York?"}]})
```

```python
response['messages'][-1].content
```

```python
# we can also invoke agent like this 
agent.invoke({"messages":"What is the weather in Boston?"})
```
