### Streaming The Responses With astream Method
#### Streaming tokens we often want to stream more than graph state.
##### In particular, with chat model calls it is common to stream the tokens as they are generated.We cand do this using the .astream_events method, which streams back events as they happen inside nodes.

##### Each event is a dict with a few keys:
* event: This is the type of event that is being emitted.
* name:  This is the name of the event
* data: This is the data associated with the event 
* metadata: Contains langgraph_node, the node emitting event.

```python
config={"configurable":{"thread_id":"3"}}
async for event in graph_builder.astream_events({"messages":["Hi,My name is Venkat and I like to play cricket"]},config,version="v2"):
    print(event)
```
