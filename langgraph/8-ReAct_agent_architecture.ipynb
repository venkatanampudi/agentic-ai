### Agent Architecture:
##### ReAct : It is general agent architecture with 3 steps in this:
1. Act     ---> Based on specific input, model calls specific tool(s)
2. Observe ---> Once the tool executes and gives the results, it will not end the conversation. Instead, it passes the tool output back to the model
3. Reason  ---> Based on the output response from the tool and input the model will decide what to do in the next step.

Here, Agent is following ReAct architecture based on the tools binded and tools defined for user queries. LLM will keep on reasoning
This is what the basic agent architecture.
