### LangGraph 
##### LangGraph is a library for building stateful, multi-actor applications with LLMs used to create agent and multi-agent workflows. LangGraph is built by LangChain Inc, but can be used without LangChain. It follows DAG (Directed Acyclic Graph)
* LangGraph powers the production grade agents
* LangGraph provides fine-grained control over both the flow and state of your agents applications. It implements a central persistence layer, enablibng features that are common to most agent architectures.
* Memory: LangGraph persists arbitrary aspects of your application's state, supporting memory of conversations and other updates within and across user interactions
* Human-in-the-loop: Because state is checkpointed, execution can be interrupted and resumed, allowing for decisions, validations, and corrections at key stage via human point.

#### LangGraph Components:
* Nodes: Every node have some kind of functionality
* Edges : The relation between nodes 
* Chatbot (multi agent ) : multi  nodes and multi edges workflow managed by LangGraph
